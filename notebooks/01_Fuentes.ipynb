```python
from pathlib import Path
import os
import pandas as pd
from ucimlrepo import fetch_ucirepo
import hashlib
import warnings
warnings.filterwarnings('ignore')
```

```python
# Rutas relativas desde el notebook
csv_path = Path("..//data/raw/data.csv")
checksum_path = Path("..//data/raw/data.csv.sha256")

# Crear carpeta si no existe
csv_path.parent.mkdir(parents=True, exist_ok=True)

# Descargar dataset desde UCI
bs = fetch_ucirepo(id=697)
df = pd.concat([bs.data.features, bs.data.targets], axis=1) # Concatenamos features y targets en un solo dataframe
df.to_csv(csv_path, index=False)

df = pd.read_csv(csv_path)
print(df.shape)
print(df.isnull().sum().sum())

# Guardar checksum para auditoría
h = hashlib.sha256()
with open(csv_path, "rb") as f:
    for chunk in iter(lambda: f.read(8192), b""):
        h.update(chunk)
checksum = h.hexdigest()
print("Checksum path:", checksum_path)
try:
    checksum_path.write_text(checksum)
    print("Checksum guardado exitosamente.")
except Exception as e:
    print("Error al guardar checksum:", e)
print("SHA256:", checksum)
```
